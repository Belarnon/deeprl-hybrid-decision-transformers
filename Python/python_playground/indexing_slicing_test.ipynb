```python
import torch
import torch.nn as nn

batch_size = 3
t = 6

# Create random tensor of shape (batch_size, t, t)
x = torch.rand(batch_size, t, t)

# Create indices tensor for the upper triangular part of the tensor
indices = torch.triu_indices(t, t, offset=1)

# Set the indices of the upper triangular part of the mask tensor to 0
x[:, indices[0], indices[1]] = 0

print(x)
print(indices[0])
print(indices[1])
print(indices.shape)
```

```python
# Test masked_fill
x = torch.rand(batch_size, t, t)

# Create a mask tensor of shape (t, t) with 0s (aka False) in the upper triangular part (excluding the diagonal)
mask = torch.tril(torch.ones(t, t))

# Set the indices of the upper triangular part of the mask tensor to 0
# x[:, mask] = 0
x = x.masked_fill(mask==0, 0)

print(x)
print(mask)
print(mask.shape)
```
